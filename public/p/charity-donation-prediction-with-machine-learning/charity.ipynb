## 2. Exploratory Data Analysis (EDA)
Analyze key variables with visualizations.

```python
# Visualize the distribution of key numerical variables


# Set the global theme for Seaborn plots
# Using "whitegrid" style for background and scaling fonts for better readability
sns.set_theme(style="whitegrid", font_scale=1.2)

# List of numerical columns to plot distributions for
numerical_columns = ['avhv', 'incm', 'inca', 'tgif', 'lgif', 'rgif', 'agif', 'damt']

# Corresponding x-axis labels for the numerical columns
x_axis_labels = [
    'average property value', 'income', 'adjusted income', 
    'total donation', 'last donation amount', 
    'latest donation amount', 'average donation amount', 
    'donation amounts'
]


# Create a blue color palette based on the number of numerical columns and this palette ensures each plot has a unique shade of blue
palette = sns.color_palette("Blues", n_colors=len(numerical_columns))

# Create a 4x2 grid layout for subplots (4 rows and 2 columns) and set the overall figure size
fig, axes = plt.subplots(4, 2, figsize=(14, 10))

# Flatten the axes to easily iterate over each subplot
axes = axes.flatten()

# Loop through each numerical column to plot a histogram
for i, column in enumerate(numerical_columns):
    # Create a histogram with a kernel density estimation (KDE) curve
    sns.histplot(
        data=train_data,           # The dataset
        x=column,                  # The column to plot
        bins=20,                   # Number of bins in the histogram
        kde=True,                  # Add a smooth KDE curve
        ax=axes[i],                # Plot on the corresponding subplot
        color=palette[i % len(palette)],  # Set the color using the palette
        alpha=0.8                  # Set transparency for better visualization
    )
    # Add a title and label the axes for each subplot
    axes[i].set_title(f'Distribution of {column}', fontsize=12)
    axes[i].set_xlabel(x_axis_labels[i], fontsize=10)
    axes[i].set_ylabel("Frequency", fontsize=10)
    # Add a grid for the y-axis to improve readability
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

# Remove any extra subplot frames if there are more subplots than columns
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Add a global title to the entire figure
fig.suptitle("Distributions of Key Numerical Features", fontsize=18, y=0.95)

# Adjust the layout of subplots for better spacing
plt.tight_layout()
plt.subplots_adjust(top=0.92)  # Leave space for the global title

plt.show()
```

```python
# Visualize the correlation matrix to observe relationships between variables

# Set the global theme for Seaborn plots
# Use a white background style and standard font scaling
sns.set_theme(style="white", font_scale=1.0)

# Calculate the correlation matrix for the selected numerical columns
# This matrix shows pairwise correlation coefficients between variables
correlation_matrix = train_data[numerical_columns].corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(6, 5))  # Set the figure size to be compact

# Generate a mask to cover the upper triangle of the heatmap
# This avoids showing duplicate values since the correlation matrix is symmetric
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Draw the heatmap
sns.heatmap(
    correlation_matrix,              # The correlation matrix to visualize
    mask=mask,                       # Apply the mask to hide the upper triangle
    annot=True,                      # Display the correlation values in the heatmap
    cmap='coolwarm',                 # Use a diverging color palette for better visualization
    fmt=".2f",                       # Format the correlation values to 2 decimal places
    annot_kws={'size': 9},           # Control the font size of the annotations
    linewidths=0.3,                  # Add thin gridlines between cells for clarity
    cbar_kws={'shrink': 0.75, 'aspect': 20}  # Customize the color bar size and aspect ratio
)

# Add a title to the heatmap with a specific font size and padding for spacing
plt.title("Correlation Heatmap", fontsize=14, pad=15)

# Rotate the x-axis labels for better readability and adjust their font size
plt.xticks(rotation=45, ha='right', fontsize=10)

# Adjust the font size for y-axis labels
plt.yticks(fontsize=10)

# Automatically adjust layout for a compact and neat plot
plt.tight_layout()

plt.show()
```

```python
# Visualize wealth rating vs. donation status

# Set the global theme for Seaborn plots
# Use "whitegrid" for a subtle grid background and standard font scaling
sns.set_theme(style="whitegrid", font_scale=1.0)

# Create a boxplot to visualize the relationship between donation status (donr) and wealth rating (wrat)
plt.figure(figsize=(6, 4))  # Set the figure size to make it compact
sns.boxplot(
    x='donr',                  # Categorical variable on the x-axis (Donation Status: Donor or Non-Donor)
    y='wrat',                  # Numerical variable on the y-axis (Wealth Rating)
    data=train_data,           # The dataset containing the variables
    palette="Set2",            # Use a soft color palette for visual appeal
    width=0.5,                 # Adjust the box width to make the plot less cluttered
    fliersize=2                # Make outlier points smaller for a cleaner look
)

# Add a title and labels to make the plot easy to interpret
plt.title('Wealth Rating by Donation Status', fontsize=12, pad=10)  # Add a title with spacing from the plot
plt.xlabel('Donation Status', fontsize=10)  # Label the x-axis
plt.ylabel('Wealth Rating', fontsize=10)  # Label the y-axis

# Customize x-axis labels to display descriptive names instead of numeric codes
plt.xticks(ticks=[0, 1], labels=['Non-Donor', 'Donor'], fontsize=9)  # Add labels for clarity

# Add horizontal gridlines to make it easier to read the Wealth Rating values
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Dashed lines with slight transparency

# Adjust layout to prevent overlap of plot elements
plt.tight_layout()


plt.show()
```

## 3. Classification Model for DONR

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Prepare training data
X_train = train_data.drop(columns=['ID', 'donr', 'damt'])  # Drop 'ID' and 'damt' as instructed
y_train = train_data['donr']

# Prepare test data
X_test = test_data.drop(columns=['ID', 'donr', 'damt'])
y_test = test_data['donr']



# Initialize and train the RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predict on the test data
y_pred = rf_model.predict(X_test)
```

```python

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Classification Accuracy for RandomForest:", accuracy)
```

```python
classification_rep = classification_report(y_test, y_pred)
classification_rep
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

## 4. Prediction (Regression) Model for DAMT (Linear Regression)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, r2_score, mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression


# Prepare data for regression model
X_train_reg = train_data[train_data['donr'] == 1].drop(columns=['ID', 'donr', 'damt'])  # Only rows with donr=1
y_train_reg = train_data[train_data['donr'] == 1]['damt']

# Prepare test data for regression model
X_test_reg = test_data[test_data['donr'] == 1].drop(columns=['ID', 'donr', 'damt'])
y_test_reg = test_data[test_data['donr'] == 1]['damt']

# Scale the features for regression
X_test_reg_scaled = scaler.fit_transform(X_test_reg)
X_train_reg_scaled = scaler.fit_transform(X_train_reg)
```

```python
# Train the regression model
linreg_model = LinearRegression()
linreg_model.fit(X_train_reg_scaled, y_train_reg)

# Predict on test data for regression model
y_pred_reg = linreg_model.predict(X_test_reg_scaled)

# Evaluate the regression model
mse_reg = mean_squared_error(y_test_reg, y_pred_reg)
mae_reg = mean_absolute_error(y_test_reg, y_pred_reg)
r2_reg = r2_score(y_test_reg, y_pred_reg)

# Display results for classification and regression models
print("R2 score for regression model: ",r2_reg)
print("Mean Squared Error (MSE) for the regrssion model: ", mse_reg)
print("Mean Absolute Error (MAE) for the regression model: ",mae_reg)
```

```python
# Adding the extracted metrics for Linear Regression
metrics = {
    "R² Score": 0.563,
    "MSE": 1.883,
    "MAE": 0.986
}

# Plotting the metrics with values annotated
plt.figure(figsize=(8, 6))
bars = plt.bar(metrics.keys(), metrics.values(), color=['green', 'blue', 'orange'], alpha=0.7)
plt.title("Linear Regression Performance Metrics")
plt.ylabel("Value")
plt.ylim(0, max(metrics.values()) * 1.2)

# Annotate the values on the bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05, 
             f"{bar.get_height():.3f}", ha='center', fontsize=10, color='black')

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Outcome of Regression Model for DAMT (Linear Regression):

Mean Squared Error (MSE): 1.883

Mean Absolute Error (MAE): 0.986

R² Score: 0.563

R2 score for regression model:  0.56, indicating that it explains about 56% of the variance in DAMT
